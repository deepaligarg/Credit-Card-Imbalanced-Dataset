```python
#Kmeans Clustering

from sklearn.cluster import KMeans
wcss=[]
    
for k in range (1,30):
    km=KMeans(n_clusters=k)
    km.fit(normdf)
    wcss.append(km.inertia_)
    
    
```

```python
wcss
```

```python
len(wcss)
```

```python
#Elbow Grapd

plt.plot(range(1,30),wcss)
plt.xlabel('K=values')
plt.ylabel('Variation(wcss)')
```

```python
#Hierarchial Clustering

from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as hcluster
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
#using Avg linkage

avg=linkage(normdf, method='average',metric='euclidean')
dendrogram(avg)
```

```python
plt.show()
```

```python
avg=linkage(normdf, method='average',metric='euclidean')
dendrogram(avg)
plt.show()
```

```python
single=linkage(normdf, method='single',metric='euclidean')
dendrogram(single)
plt.show()
```

```python
compl=linkage(normdf, method='complete',metric='euclidean')
dendrogram(compl)
plt.show()
```
